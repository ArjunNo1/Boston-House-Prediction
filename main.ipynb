```python
from sklearn.impute import SimpleImputer

imputer = SimpleImputer(strategy="median")
df[['CRIM', 'ZN', 'INDUS', 'CHAS','AGE','LSTAT']] = imputer.fit_transform(df[['CRIM', 'ZN', 'INDUS', 'CHAS','AGE','LSTAT']])

df.isnull().sum()
```

```python
df = df.astype({'RAD':'float64','TAX':'float64'})
df.info()
```

```python
# Standardize ( Scale ) the data so that if all the features are having same unit ,then  the gradient will converge at Global minima fast
```

```python
df.head()
```

```python
# To know how data is distributed along each column 

plt.figure(figsize=(10,10))
plot = 1

for feature in df:
    if plot<=14:
        ax = plt.subplot(3,5,plot)
        sns.distplot(df[feature])
        plt.xlabel(feature,fontsize = 14)
    plot+=1
plt.tight_layout()
plt.show()
```

```python
# Comparing our target Feature MEDV (Price) with rest of Independent Features

plt.figure(figsize=(10,10))
plot = 1

for feature in df:
    if plot<=14:
        ax = plt.subplot(3,5,plot)
        sns.scatterplot(x=df['MEDV'],y=df[feature])
    plot+=1
plt.tight_layout()
plt.show()
        
```

```python
# Seeing for Outliers -Boxplot

plt.figure(figsize=(20,9))
sns.boxplot(data=df,width=0.8)
plt.show()
```

```python
# Here we can Standardise our data so that all the features have common unit .This can end up leading the gradient to converge at global maxima Fast

X=df.drop(columns='MEDV',axis=1)
y=df['MEDV']
```

```python
from sklearn.preprocessing import StandardScaler

scaler  = StandardScaler()
scaled_X = scaler.fit_transform(X)
```

```python
# Checking for Multicollinearity using `Variance Influence Factor


from statsmodels.stats.outliers_influence import variance_inflation_factor

vif = pd.DataFrame()
vif['FEATURES'] = X.columns
vif['VIF'] = [variance_inflation_factor(scaled_X,column) for column in range(scaled_X.shape[1])]

vif
```

## We could see that feature TAX and RAD having high correlation, So it is good to drop any of these


## The smf module in statsmodels (short for "statsmodels.formula.api") is a part of the statsmodels library in Python that provides a formula-based approach to specifying statistical models. The smf module is built on top of the statsmodels library and provides an interface for specifying models using R-style formulas.

## With smf, you can easily specify regression models, time series models, mixed-effect models, and other types of statistical models using a simple formula notation. The formula notation allows you to specify the dependent and independent variables of the model and their relationships using a formula syntax.

## For example, to fit a linear regression model with one independent variable, you can use the ols function from smf 

```python
import statsmodels.formula.api as smf

lm = smf.ols(formula='MEDV~RAD',data=df).fit()

lm.summary()
```

```python
lm = smf.ols(formula='MEDV~TAX',data=df).fit()

lm.summary()
```

```python
# We can drop the RAD column since it has less condition number in the above summary

df.drop(columns='RAD',axis=1,inplace=True)
```

```python
df.head()
```

```python
from sklearn.model_selection import train_test_split

xtrain,xtest,ytrain,ytest = train_test_split(scaled_X,y,test_size=0.30,random_state = 42)
```

```python
from sklearn.linear_model import LinearRegression

model = LinearRegression()
model.fit(xtrain,ytrain)
```

```python
prediction = model.predict(xtest)
```

```python
print(model.coef_)
print(model.intercept_)
```

```python
from sklearn.metrics import mean_absolute_error , mean_squared_error,r2_score

print(mean_absolute_error(ytest,prediction))
print(mean_squared_error(ytest,prediction))
print(r2_score(ytest,prediction))
```

```python
print(model.score(xtest,ytest))   # Model accuracy on new Data

print(model.score(xtrain,ytrain))  #Model accuracy on Trained Data
```

```python
# Adjusted R2-Score
#n-->no.of.observations
#p-->no.of.features

def adjusted_r2(X,y,m):
    r2 = model.score(X,y)
    n = X.shape[0]
    p =X.shape[1]
    adj_r2_score = 1-(1-r2)*(n-1) / (n-p-1)
    return adj_r2_score

# print(adjusted_r2(xtrain,ytrain,model))

print(adjusted_r2(xtest,ytest,model))

```
